```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
# here in clustering there is no Y. X,Y is supervised
```

```python
X = dataset.iloc[:,[3,4]].values # take all rows and 3rd,4th column
```

```python
X
```

```python
from sklearn.cluster import SpectralClustering
```

```python
SC = SpectralClustering(n_clusters=5, assign_labels='discretize',random_state=0).fit(X)
y_SC = SC.fit_predict(X)
```

```python
y_SC
```

```python
supervised = pd.DataFrame(dataset)
```

```python
supervised['Cluster_group'] = y_SC # adding extra column to dataset
```

```python
supervised
```

```python
import seaborn as sns # graphical purposes
```

```python
# 'legen'd gives the color info, 'legend_out' places it side
```

```python
facet = sns.lmplot(data=supervised, x = supervised.columns[3], y = supervised.columns[4], 
                 hue = supervised.columns[5], fit_reg = False, legend = True, legend_out = True )
```
